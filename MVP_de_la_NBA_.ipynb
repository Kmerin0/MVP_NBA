```python
candidatos['%puntos'] = (candidatos['Puntos_totales']/2225) * 0.25
candidatos['%asistencias'] = (candidatos['Asistencias']/ 678) * 0.25
candidatos['%rebotes'] = (candidatos['Rebotes_totales']/ 817) * 0.25
candidatos['%juegos'] = (candidatos['Juegos']/83) * 0.25
candidatos['MVP_candidato'] = candidatos['%puntos'] + candidatos['%asistencias'] + candidatos['%rebotes'] + candidatos['%juegos']
#candidatos.sort_values('MVP_candidato', ascending = False)




```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Crear el dash
app = dash.Dash(__name__)


# Obtener las columnas dentro de mi dataframe.
menu_drop = candidatos.loc[:,['%puntos', '%asistencias',	'%rebotes',	'%juegos','MVP_candidato']]

# Definir el layout de la aplicación

app.layout = html.Div([
    html.Div([
        html.H1('MVP de la NBA'),
        html.Div([
            dcc.Dropdown(
                id = 'yaxis-column',
                options = [{'label': i, 'value': i} for i in menu_drop],
                value = '%juegos'
            ),
            dcc.Graph(id = 'indicator-graphic')
        ],
        style={'width': '100%', 'display': 'inline-block'})
    ])
], style={'background-color': '#000000'})

# Definir la función de callback para actualizar la gráfica
@app.callback(
    Output('indicator-graphic', 'figure'),
    Input('yaxis-column', 'value')
)
def update_graph(yaxis_column_name):
    # Crear el gráfico con los datos seleccionados
    fig = px.bar(candidatos, x='Jugador', y=yaxis_column_name, hover_data=[yaxis_column_name])

    # Personalizar el diseño del gráfico
    fig.update_layout(margin={'l': 40, 'b': 40, 't': 10, 'r': 0}, hovermode='closest')
    fig.update_xaxes(title='Jugador')
    fig.update_yaxes(title=yaxis_column_name)

    return fig

# Ejecutar la aplicación Dash
if __name__ == '__main__':
    app.run_server(debug=True)
```
